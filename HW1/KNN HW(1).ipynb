```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def averageAcc(average_acc_per_k):
    averageIn = sum(average_acc_per_k)/5
    return averageIn

def main():
    
    plt.style.use('seaborn-whitegrid')
    file = 'iris.data'
    col_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
    df = pd.read_csv(file, header = None, names = col_names)
    iris_class = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
    df['species_num'] = [iris_class[i] for i in df.species]
    X = df.drop(['species', 'species_num'], axis=1)
    y = df.species_num
    accuracy = []
    average_acc_per_k = []
    #start loop to insert values and test n_neighbors here
    for i in range (1, 21):
        for j in range(1, 6):
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .34)
            k = KNeighborsClassifier(n_neighbors=i)
            k.fit(X_train, y_train)
            y_pred = k.predict(X_test)
            accuracy.append(metrics.accuracy_score(y_test, y_pred))
        
        average_acc_per_k.append(averageAcc(accuracy))
        
        accuracy.clear()
        
    
    print('\nAverage Accuracy per K-value: ' + repr(average_acc_per_k))
    
    fig = plt.figure()
    ax = plt.axes()
    
    plt.title("Accuracy model")
    plt.xlabel("K-Values")
    plt.ylabel("Average Accuracy")
    
   
    ax.plot(np.array(average_acc_per_k))
    
    
main()

```
